```python
import numpy as np
import pandas as pd
```

```python
smartphones = pd.read_csv('Flipkart Mobile - 2.csv', encoding = ('ISO-8859-1'), low_memory = False)

```

```python
smartphones.head()
```

```python
smartphones.info()
```

```python
smartphones['sales_price']=str(smartphones['sales_price'])
smartphones['battery_capacity']=str(smartphones['battery_capacity'])
```

```python
smartphones.info()
```

```python
smartphones['tags']=smartphones['brand']+ " " +smartphones['model']+ " " +smartphones['processor']+ " " +smartphones['screen_size']+ " "+ smartphones['base_color']
```

```python
smartphones['tags'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=4300)
```

```python
vectors = cv.fit_transform(smartphones['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[408])), reverse = True,key = lambda x:x[1])[1:6]
```

```python
def recommend(smartphone):
    smartphone_index = smartphones[smartphones['model'] == smartphone].index[0]
    distances = similarity[smartphone_index]
    smartphones_list = sorted(list(enumerate(distances)), reverse = True,key = lambda x:x[1])[1:6]
    
    for i in smartphones_list:
        print(smartphones.iloc[i[0]])
        
    
    
    
```
